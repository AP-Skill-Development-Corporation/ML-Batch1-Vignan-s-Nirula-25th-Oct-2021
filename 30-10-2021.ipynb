### DMR(Dimentionality Reduction):

* Reduce the no.of dimentions of data
* It improves the model performance

* PCA(Principle component of analysis):
    * scale the data
    * fit the scale data into pca
    * convert pca data into dataframe


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
d=load_boston()
df=pd.DataFrame(d.data,columns=d.feature_names)
df['target']=d.target
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x=df.iloc[:,:-1]
a=s.fit_transform(x)
a
```

```python
from sklearn.decomposition import PCA
p=PCA(n_components=4)
pc=p.fit_transform(a)
p_df=pd.DataFrame(pc,columns=["p1","p2",'p3','p4'])
p_df
```

```python
x=p_df[['p1','p2','p3','p4']]
y=df['target']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
model.score(x,y)
```

### cross validation

```python
from sklearn.datasets import load_boston
d=load_boston()
df=pd.DataFrame(d.data,columns=d.feature_names)
df['target']=d.target
df.head()
```

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.model_selection import cross_val_score,KFold
kfold=KFold(n_splits=5,shuffle=True)
model=LinearRegression()
c=cross_val_score(model,x,y,cv=kfold,scoring='r2')
c
```
